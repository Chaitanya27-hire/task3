```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris = pd.read_csv(r'C:\Users\AKASH\Desktop\pro3\Iris.csv')
```

```python
iris.head()
```

```python
#select the feature you want to visulize
feature_to_plot = "SepalLengthCm"
```

```python
#create a histogram sepalLengthCm
plt.figure(figsize=(8, 6))
sns.histplot(iris[feature_to_plot], kde=True, color="blue")
plt.title(f"Histogram of {feature_to_plot} in Iris Dataset")
plt.xlabel(feature_to_plot)
plt.ylabel("Frequency")
plt.show()
```
